```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

### Neural Network performance on antibody-escape phenotype for 10 human monoclonal antibodies (Figure 6B; Table S3)

```python
d_Abs = {'COV2-2050': [0.246994], 'COV2-2082': [0.09224], 'COV2-2094': [0.315764], 'COV2-2096': [0.19175], 'COV2-2165': [0.123021], 'COV2-2479': [0.130445], 'COV2-2499': [0.260286], 'COV2-2677': [0.29153], 'COV2-2832': [0.125652], 'rCR3022': [0.264203]}
df_perf = pd.DataFrame(data=d_Abs, index=["Performance"])
```

```python
df_perf.iloc[0,]
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(df_perf.columns, df_perf.iloc[0,],hue=df_perf.columns,legend = False)
plt.ylim(0,0.4)
plt.xticks(rotation=45)
plt.title("Neural network model performance", fontsize=16)
plt.ylabel("Spearman correlation coefficient",fontsize=16)
plt.show()
```
